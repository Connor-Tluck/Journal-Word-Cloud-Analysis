```python
import os
import textract
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from collections import Counter

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#scan through documnets
filelist = []

directory = '/Users/connortluck/Desktop/Word_Cloud_Depression/Entries/'
for entry in os.scandir(directory):
    if (entry.path.endswith(".docx")
            or entry.path.endswith(".doc")) and entry.is_file():
        filelist.append(entry.path)

#read through each entry and decode to text then append to a large string 
total_text = ''
for filepath in filelist:
    text = textract.process(filepath)
    decoded_text = text.decode('utf-8')
    total_text += decoded_text.lower()
# print(total_text[:10000])

# image = Image.open("/Users/connortluck/Desktop/Archive/Solid_black.png")
# new_image = image.resize((1000, 1000))
# new_image.save("/Users/connortluck/Desktop/Archive/Solid_black_LS.png")


comment_words = total_text
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in total_text: 
      
    # typecaste each val to string 
    val = str(val) 

# split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
        
        comment_words += " ".join(tokens)+" "


wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (9, 9), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
# to recolour the image
char_mask = np.array(Image.open("/Users/connortluck/Desktop/Archive/Solid_black_LS.png"))    
image_colors = ImageColorGenerator(char_mask)

plt.imshow(wordcloud.recolor(color_func=image_colors))

# to recolour the image
char_mask = np.array(Image.open("/Users/connortluck/Desktop/Archive/Solid_black_LS.png"))    
image_colors = ImageColorGenerator(char_mask)

plt.imshow(wordcloud.recolor(color_func=image_colors))

```
